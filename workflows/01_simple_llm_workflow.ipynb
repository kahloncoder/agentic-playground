```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python
 model = ChatOpenAI() 
```

```python
class llmstate(TypedDict):
    input: str
    output: str
```

```python
def llm_qe(state: llmstate) -> llmstate:
    question = state['input']
    prompt = f"provide a concise answer to the following question: {question}"
    answer = model.invoke(prompt).content

    state['output'] = answer
    return state
```

```python
graph = StateGraph(llmstate)

#add node
graph.add_node('llm_qe', llm_qe)

#add edge
graph.add_edge(START, 'llm_qe')
graph.add_edge('llm_qe', END)

workflow = graph.compile()

```

```python
initial_state = {"input": "What is LangGraph?"}

final_state= workflow.invoke(initial_state)

print(final_state['output'])
```
