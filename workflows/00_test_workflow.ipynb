```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict
```

```python
class testState(TypedDict):
    value_x: float
    value_y: float
    test: float
```

```python
def claculate(state: testState) -> testState:
    X=state['value_x']
    Y=state['value_y']
    value= X*Y
    state['test']=round(value,2)
    return state
```

```python
#define the graph
graph = StateGraph(testState)

#add nodes to the graph
graph.add_node('claculate', claculate)

#add edges to the graph
graph.add_edge(START, 'claculate')
graph.add_edge('claculate', END)

#comile the graph
workflow =graph.compile()
```

```python
#execute the graph
workflow.invoke({'value_x': 21, 'value_y': 2})
```

```python
Initialstate={'value_x': 21, 'value_y': 2}

finalstate=workflow.invoke(Initialstate)

print(finalstate)
```
